### Read DB

```python
import os
from supabase import create_client, Client
url: str = os.environ.get("SUPABASE_URL")
key: str = os.environ.get("SUPABASE_KEY")
username: str = os.environ.get("SUPABASE_EMAIL")
password: str = os.environ.get("SUPABASE_PWD")

supabase: Client = create_client(url, key)
auth = supabase.auth.sign_in_with_password(
    {
        "email": username,
        "password": password,
    }
)

# Get all table
result = supabase.rpc("get_tables").execute()
print('All created table')
print(result.data)
```

### Edit Table

```python
# Insert data
response = (
    supabase.table("USERS")
    .insert({"USER_ID": "1", "USERNAME": "test", "CREATED_AT":"2025-08-12", "ROLE": "driver"})
    .execute()
)
print('created', response)
```

```python
# Delete data
response = (
    supabase.table("USERS")
    .delete()
    .eq("USERNAME", "test")
    .execute()
)
print('deleted', response)
```
